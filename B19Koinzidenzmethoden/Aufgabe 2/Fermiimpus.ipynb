```python
#Methoden Parabelfit f√ºr Mittelwerte
def parabola(x, a, b, c):
    return a*x**2 + b*x + c

def parabola_fit(x, y):
    # mean = sum(x * y) / sum(y)
    # sigma = np.sqrt(sum(y * (x - mean) ** 2) / sum(y))
    popt, pcov = curve_fit(parabola, x, y)#, p0=[min(y), max(y), mean, sigma])
    return popt[0], popt[1], popt[2], pcov[2,2]

def midnightformulaplus(a,b,c):
    return (-b+np.sqrt(b**2-4*a*c))/(2*a)

def midnightformulaminus(a,b,c):
    return (-b-np.sqrt(b**2-4*a*c))/(2*a)

#Fit
afit, bfit, cfit, fehlerm√º= parabola_fit(deltaX, Nsums)
print('MW:',round(afit,3), '$\pm$', np.sqrt(fehlerm√º))

NSminus = midnightformulaminus(afit,bfit,cfit)
NSplus = midnightformulaplus(afit,bfit,cfit)
# NSfehler = midnightformulaplus(afit,bfit,cfit+np.sqrt(parabola(NSplus, afit,bfit,cfit)))
# print("NSFehler:",NSfehler)
print(NSminus)
print(NSplus)

maxdeltax = (NSminus-NSplus)/2
print(maxdeltax)
l = 0.7#m
py = 511000

pF = (maxdeltax*10**(-3)*511000)/l
print("pF=",pF)

Ef = (pF**2)/(2*511000)
print("Ef=",Ef)

```
